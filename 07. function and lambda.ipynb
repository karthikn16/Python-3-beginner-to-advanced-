```python
#functions
def my_function(fname):
  print(fname + " python")

my_function("john")
my_function("ram")
my_function("bala")
```

```python
# two parameters
def my_function(fname, lname):
  print(fname + " " + lname)

my_function("tom", "john")
```

```python
#arbitary arguments

def my_function(*kids):
  print("The youngest child is " + kids[2])

my_function("Elil", "bob", "nike")
```

```python
#list in argument
def my_function(food):
  for x in food:
    print(x)

fruits = ["apple", "banana", "cherry"]

my_function(fruits)
```

```python
#lambda

x = lambda a : a + 10
print(x(5))
```

```python
x = lambda a, b : a * b
print(x(5, 6))
```

```python
def myfunc(n):
  return lambda a : a * n

mydoubler = myfunc(2)
mytripler = myfunc(3)

print(mydoubler(15))
print(mytripler(10))
```
