```python
#if statement
a = 20 
b = 50 
if a < b :
  print ( " a is less than b" )
```

```python
#elseif

c = 10 
d = 10
if c > d:
  print ("c is greater than d")
elif c==d :
  print ("c is equal to d")
```

```python
#nested if statement

t = 20
if t >10 :
  print("t is greater than 10")
  if t >15:
    print("and also greater than 15")
  
```
