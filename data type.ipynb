```python
#text type
a = "hello world"
print(a)
print(type(a))
```

```python
#numeric type
#int , #float ,#complex

#int

a = 25
print(a)
print(type(a))
```

```python
#float

a = 2.4
print(a)
print(type(a))
```

```python
#complex

a = 5j
print(a)
print(type(a))
```

```python
#sequence data type
#list ,#tuples , #range

#list

b = ["green","orange","red"]
print(b)
print(type(b))
```

```python
#tuples

c = ("green","red" ,"orange")
print(c)
print(type(c))
```

```python
#range

d = range(10)
print(d)
print(type(d))
```

```python
#dict

a = {"car":"bike","train": 3}
print(a)
print(type(a))
```

```python
#mapping type 
#set
a = {"red","green","orange"}
print(a)
print(type(a))
```

```python
#boolen type
l = True 
m = False
print (l)
print (m)
print(type(l))
print(type(m))
```

```python
#binary type

#bytes

a = b"HI"
print(a)
print(type(a))
```

```python
#byte array
o = bytearray (2)
print(bytearray(o))
print(type(o))
```

```python
#memoryview

f = memoryview(bytes(5))
print(memoryview(f))
print(type(f))
```
